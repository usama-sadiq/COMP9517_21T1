```python
# func to calculate the hog feature of the img
# return the features and hog img

def hog_features(img, orient, pixels_per_cell, cells_per_block, show_img = False, show_feature_vec = True):
    
    #if we also want to show the hog img along with hog features
    if show_img == True:
        
        features, hog_img = hog(img, orientations = orient, pixels_per_cell = (pixels_per_cell, pixels_per_cell), cells_per_block = (cells_per_block, cells_per_block), transform_sqrt = False, visualize = show_img)
        
        return features, hog_img
        
    else:
        
        features = hog(img, orientations = orient, pixels_per_cell = (pixels_per_cell, pixels_per_cell), cells_per_block=(cells_per_block, cells_per_block), transform_sqrt = False, visualize = show_img)
        
        return features
```

## Extract the Hog features

### I tired 2 methods

### 1) I calculated the hog features of all the channels combined but that didnt give me good results

### 2) Extract the hog features of a single channel and then combine them which gave me better results

```python
#func to extract all the hog features

def extract_hog_features(imgs, orient = 9, pixels_per_cell = 16, cells_per_block = 1):
    
    #build a list of the hog features for all the images
    features = []
    
    #Looping through the images
    for i in imgs:
        
        #Using the YCrCb as it gave me the best results
        #Convert the img to YCrCb
        feature_image = cv.cvtColor(i, cv.COLOR_BGR2YCrCb)
        
        #Hog features of the image. Extracted separatly and then combined
        hog_features_L = []
        
        for channel in range(feature_image.shape[2]):
            
            hog_features_L.append(hog_features(feature_image[:,:,channel],
                                     orient, pixels_per_cell, cells_per_block, show_img=False, show_feature_vec=True))
        
        
        #shape of Hog_Features_L C * Hog Features of the Color channel
        # 3 * 81
        
        #Convert the hog features to 1 D array
        hog_features_L = np.ravel(hog_features_L)
        
        #print(hog_features_L.shape)
        
        #features_hog = hog_features(feature_img, orient, pix_per_cell, cell_per_block, visual_img=False, feature_vec = True)
        
        #Append the hog features of the image 
        features.append(hog_features_L)
        
        
    
    return features
```

### Implementing sliding window

#### We will implement sliding window as we want to slide a window over the image and using our classifier predict if it is a car or not and using that we can make a bounding box

```python
def sliding_window(img, xlim, ylim, xy_WindowSize, xy_Overlap):
    
    #print(img.shape, xlim, ylim, xy_Window_Size, xy_Overlap)
    
    #Declare the default values for xlim and ylim if no values given
    
    if xlim == [None, None]:
        
        xlim[0] = 0
        xlim[1] = img.shape[1]
        
        #print(xlim)
    
    if ylim == [None, None]:
        
        ylim[0] = 0
        ylim[1] = img.shape[0]
    
    #print(xlim[1], xlim[0])
    
    #print(ylim[1], ylim[0])
    
    #Now we have to calculate the area/region which we want to be searched
    deltaX = xlim[1] - xlim[0]
    deltaY = ylim[1] - ylim[0]
    
    #print(deltaX,deltaY)
    
#   #print(xy_Overlap)
    
    #Now we have to compute the number of pixels in X axis
    no_of_pixels_per_step_x = np.int(xy_WindowSize[0] * (1 - xy_Overlap[0]))
    
    #Now we have to calculate the number of pixels in y axis
    no_of_pixels_per_step_y = np.int(xy_WindowSize[1] * (1 - xy_Overlap[1]))
    
    #print(no_of_pixels_per_step_x, no_of_pixels_per_step_y)
    
    #Now we will calculate the number of windows for X axis
    no_of_windows_x = np.int( (deltaX - ( np.int(xy_WindowSize[0] * (xy_Overlap[0] ))) ) / no_of_pixels_per_step_x)
    
    #Now we will calculate the number of windows for y axis
    no_of_windows_y = np.int( (deltaY - ( np.int(xy_WindowSize[1] * (xy_Overlap[1] ))) ) / no_of_pixels_per_step_y )
    
    #print(no_of_windows_x, no_of_windows_y)
    
    #list for appending the windows coordinates
    window_List = []
    
    for y in range(no_of_windows_y):
        
        for x in range(no_of_windows_x):
            
            #starting coord for x
            start_X = x * no_of_pixels_per_step_x + xlim[0]
            
            #ending coord for x
            end_X = start_X + xy_WindowSize[0]
            
            #starting coord for y
            start_Y = y * no_of_pixels_per_step_y + ylim[0]
                             
            #ending coord for y
            end_Y = start_Y + xy_WindowSize[1]
            
            window_List.append(((start_X, start_Y), (end_X, end_Y)))

    return window_List
```

```python
# func to search for cars using our classifier in each windows found by sliding windows and if found then append it to a list
def search_window_for_cars(img, windows_found_1, classifier, X_Scalar, orient, pixels_per_cell, cells_per_block):
    
    #store car windows
    car_found_windows = []
    
    #traverse all the images
    #extract the img of size 60,60
    #classify it
    #and if it is a car include it in the list
    
    for window_1 in windows_found_1:
        
        #setting the dim
        dim = (60,60)
        
        #setting the x_start for the window
        x_start = window_1[0][1]
        
        #setting the x_end for the window
        x_end = window_1[1][1]
        
        #setting the y_start for the window
        y_start = window_1[0][0]
        
        #setting the y_stop for the window
        y_end = window_1[1][0]
        
        #plt.imshow(img)
        
        #plt.show()
        
        #extracting the window from the original img and resizing it to 60 * 60
        img_2 = cv.resize(img[x_start:x_end, y_start:y_end ], dim)
        
        #print(img.shape)
        
        #find the hog features for the extracted img
        img_hog_features = extract_hog_features([img_2], orient=orient, pixels_per_cell=pixels_per_cell, cells_per_block=cells_per_block)
        
        #Scale the extracted features
        Scaled_img_hog_features = X_Scalar.transform(np.array(img_hog_features).reshape(1,-1))
        
        #make prediction using the classifier and the scaled hog features
        pred = classifier.predict(Scaled_img_hog_features)
        
        #if a car is found, append it to the list
        if pred == 1:
            car_found_windows.append(window_1)
        
    return car_found_windows
```

## Use the heatmap approach

### to reduce the false positive bounding boxes

```python
#func for drawing the bodunding boxes after labeling the img using heatmap
def draw_bounding_boxes_labels(img, labels):
    
    #loop through all the detected cars and draw bounding boxes for them
    
    no_of_cars_detected = labels[1] + 1
    
    for car in range(1, no_of_cars_detected):
        
        #Find all the pixel values using the car label and car value
        # and make sure they are nonzero
        
        non_zero_pixels = (labels[0] == car).nonzero()
        
        #Find the x coords of these non zero pixel.
        non_zero_pixels_X = np.array(non_zero_pixels[1])
        
        #Find the y coords of these non zero pixel.
        non_zero_pixels_Y = np.array(non_zero_pixels[0])
        
        #we will define the bounding box on the basis of the min and max values of x and y. As this will help us to decrease our false postives
        bounding_box = ((np.min(non_zero_pixels_X), np.min(non_zero_pixels_Y)), (np.max(non_zero_pixels_X), np.max(non_zero_pixels_Y)))
        
        cv.rectangle(img, bounding_box[0], bounding_box[1], (0,0,255), 6)
        
        return img


#func to generate a heat map image of the image so that we can highlight only the pixels which are pixels of the detected cars
def add_heat_to_bounding_box_pixels(heat_map, bounding_box_list):
    
    #loop through the list of bounding box and add 1 for all the pixels inside the bounding boxes
    
    for bounding_box in bounding_box_list:
        
        #We assume that each bounding boxes are of the form ((x1, y1), (x2,y2))
        
        x1 = bounding_box[0][1]
        
        y1 = bounding_box[1][1]
        
        x2 = bounding_box[0][0]
        
        y2 = bounding_box[1][0]
        
        heat_map[x1 : y1, x2 : y2] += 1
    
    return heat_map
    
#func to apply a threshold so that can we reduce our false positives in our images    
def apply_heatmap_threshold(heat_map, threshold):
    
    heat_map[heat_map <= threshold] = 0
    
    return heat_map
```

## Main Method

```python
#Input paths to the image files
Input_Img_files = ['Data\\benchmark_velocity_train\\clips\\1\\imgs\\040.jpg']

#Loop through all the input image files

for img in Input_Img_files:

    #read an image
    image = cv.imread(img)
    
    #make a copy of the image to work on
    draw_image = np.copy(image)
    
    #Initialize an image for generating a heatmap image
    heat_map_img = np.zeros_like(image[:,:,0]).astype(np.float)
    
    #Declare the values of xy_Windows, y_lim and xy_Overlap
    xy_Windows = [(86,59)]
    
    y_lim = [[341,401]]
    
    xy_Overlap = [(0.81,0.81)]
    
    #Build a list of windows using sliding window func
    windows_temp_1 = []

    #Loop through all the windows
    for i in range(len(xy_Windows)):
    
        #store the windows to search for cars
        windows_1 = sliding_window(image, [None, None], y_lim[i], xy_Windows[i], xy_Overlap[i])
        
        #append the windows to a list
        windows_temp_1.append(windows_1)

    #Flatten the windows list to 1-D list
    windows_final_1 = sum(windows_temp_1, [])
    
    #print(windows_final_1)
    
    #Declare values for orient, pixels_per_cell and cells_per_block
    
    orient = 9 #HOG orientations
    
    pixels_per_cell = 16 #HOG pixels per cell
    
    cells_per_block = 1 #HOG cells per block
    
    #search for cars in the windows one by one
    detected_car_windows = search_window_for_cars(image, windows_final_1, svc, X_Scaler, orient, pixels_per_cell, cells_per_block)
    
    #print(detected_car_windows)
    
    #Draw the detected cars on the image
    detected_car_img = draw_bounding_box(draw_image, detected_car_windows, color=(0, 0, 255), thickness=6)

    #Generate a heat_map of the image so that we can reduce the false positives in the #image
    heat_map_img = add_heat_to_bounding_box_pixels(heat_map_img, detected_car_windows)
    
    #We will apply a threshold so that we can get our final image
    heat_map_img = apply_heatmap_threshold(heat_map_img, 2)
    
    #Visualize the heatmap image. Apply clip to bring it in range so that we can #visualize it
    heatmap_img = np.clip(heat_map_img, 0, 255)
    
    #Label the image so that we can generate our final image with reduced false positives
    labels_img = label(heatmap_img)
    
    #Generate the final image of the dectected cars
    draw_image = draw_bounding_boxes_labels(np.copy(image), labels_img)
    
    fig = plt.figure(figsize=(10,10))
    
    plt.subplot(131)
    
    plt.imshow(detected_car_img)
    
    plt.title("Cars detected")
    
    plt.subplot(132)
    
    plt.imshow(heatmap_img, cmap='hot')
    
    plt.title('Heat Map image')
    
    plt.subplot(133)
    
    plt.imshow(draw_image, cmap='hot')
    
    plt.title('Final Image')
    
    plt.tight_layout()
    
    
```
